```python
def load_dataset(train_dataset='train_data.csv', test_dataset='test_data.csv'):
    """
    Loads and returns train and test datasets.
    """
    data_train = np.genfromtxt(fname=train_dataset, delimiter=',')
    y_train = data_train[:,2500:].astype(np.int)
    x_train = (data_train[:,:2500]/ 255).astype(np.int)
    data_test = np.genfromtxt(fname=test_dataset, delimiter=',')
    y_test = data_test[:,2500:].astype(np.int)
    x_test = (data_test[:,:2500]/ 255).astype(np.int)
    return (x_train, y_train, x_test, y_test)
x_train, y_train, x_test, y_test = load_dataset()
```
